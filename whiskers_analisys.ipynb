```python
%matplotlib notebook
data = pd.read_csv('1-9.csv', sep=';')
data['d_av (um)'] = (data['d1 (um)']+data['d2 (um)']+data['d_mid (um)'])/3 
data['disp'] =  np.sqrt(((data['d1 (um)']-data['d_av (um)'])**2+(data['d2 (um)']-data['d_av (um)'])**2
                    +(data['d_mid (um)']-data['d_av (um)'])**2)/3)
np.round(data, 3)
data
```
